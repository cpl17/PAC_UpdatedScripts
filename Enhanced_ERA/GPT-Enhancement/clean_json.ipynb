```python
import json

# Specify the path to your JSON file
json_file_path = 'plants_trimmed.json'

data_dictionary = {}
# Read the JSON file and load it into a dictionary
with open(json_file_path, 'r') as json_file:
    for line in json_file.readlines():
        plant_dictionary = json.loads(line)
        data_dictionary[plant_dictionary["_id"]] = plant_dictionary

# Now, 'data_dictionary' contains the contents of the JSON file as a Python dictionary
print(data_dictionary["Abies amabilis"])
```

```python
data = pd.read_csv("Blurbs_New.csv")
data = data[:10]
names,blurbs = data["Scientific Name"],data["Blurb"]

for name,blurb in list(zip(names,blurbs)):
    data_dictionary[name]["Blurb"] = blurb
```

```python
for key in data_dictionary.keys():
    print(data_dictionary[key]["Blurb"])
    
```

```python
keys_to_drop = ["_id","Local Count","Online Counts","Workhorse"]
```

```python
for name in names:
    for key in keys_to_drop:
        del data_dictionary[name][key]
```

```python
keys = [key for key in data_dictionary.keys()]
for key in keys:
    if data_dictionary[key]["metadata"] is None:
        continue
    print(data_dictionary[key]["metadata"])
    data_dictionary["metadata_extmetadata_Artist_value"] = data_dictionary[key]["metadata"]["extmetadata"]["Artist"]["value"]
    data_dictionary["metadata_extmetadata_Artist_source"] = data_dictionary[key]["metadata"]["extmetadata"]["Artist"]["source"]
    data_dictionary["metadata_extmetadata_LicenseShortName_source"] = data_dictionary[key]["metadata"]["extmetadata"]["LicenseShortName"]["source"]
    data_dictionary["metadata_extmetadata_LicenseShortName_value"] = data_dictionary[key]["metadata"]["extmetadata"]["LicenseShortName"]["value"]
```
